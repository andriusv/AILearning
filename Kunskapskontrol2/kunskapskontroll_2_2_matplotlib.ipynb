### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

```python
Figure
The Figure is the entire window or canvas that contains all plots.

Axes
The Axes is the area where data is plotted. A figure can have multiple axes.

Axis
An Axis is a single coordinate axis (x or y) that controls scale, ticks, and labels.

Artist
An Artist is any visible element in a plot, such as lines, text, or bars.
```

### When plotting in Matplotlib, what is the expected input data type?

```python
Numeric data as lists, tuples, NumPy arrays, or Pandas Series
```

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
import numpy as np;
import matplotlib.pyplot as plt 
x = np.linspace(-4, 4, 100)
y = x**2

plt.plot(x, y)
plt.title("y = x²")
plt.xlabel("x")
plt.ylabel("y")
plt.show()


x = np.linspace(-4, 4, 100)
y = x**2

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title("y = x²")
ax.set_xlabel("x")
ax.set_ylabel("y")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

ax1.scatter(random_data_x, random_data_y, color='blue', alpha=0.6)
ax1.set_title("Random Scatter Plot")
ax1.set_xlabel("X values")
ax1.set_ylabel("Y values")

# Bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())

ax2.bar(names, values, color='orange')
ax2.set_title("Fruit Counts")
ax2.set_xlabel("Fruit")
ax2.set_ylabel("Count")

plt.tight_layout()
plt.show()
```
